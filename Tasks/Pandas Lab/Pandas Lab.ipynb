```python
'''
3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]
'''
even = pd.Series([2, 4, 6, 8, 10])
odd = pd.Series([1, 3, 5, 7, 9])
arthdf = pd.DataFrame({'Even':even, 'Odd':odd})
arthdf['Add'] = arthdf['Even'] + arthdf['Odd']
arthdf['Sub'] = arthdf['Even'] - arthdf['Odd']
arthdf['Mul'] = arthdf['Even'] * arthdf['Odd']
arthdf['Div'] = arthdf['Even'] / arthdf['Odd']
arthdf
```

```python
'''
4. Selecting all the rows from the given dataframe in which ‘Percentage’ is greater than 80
using basic method.
'''
record = {
'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ], 'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math','Science'], 'Percentage': [88, 92, 95, 70, 65, 78] }
columns = ['Name', 'Age', 'Stream', 'Percentage']

df = pd.DataFrame(data = record, columns = columns)
df[df['Percentage'] > 80]
```
