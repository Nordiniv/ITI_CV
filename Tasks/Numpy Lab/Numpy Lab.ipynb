```python
'''
3- Write a NumPy program to append values to the end of an array.
Expected Output:
Original array:
[10, 20, 30]
'''

np.hstack([np.array([10, 20, 30]), np.array([40, 50, 60, 70, 80, 90])])
```

```python
# 4- Write a NumPy program to create a 3X4 array and iterate over it using nditer.

arr3x4 = np.arange(12).reshape(3, 4)

for x in np.nditer(arr3x4):
    print(x)
```

```python
'''
5- Write a NumPy program to create a 1-D array then convert it into 2-
D array then print value and position for each value using
ndenumerate.
'''

for idx, val in np.ndenumerate(arr3x4):
    print(f'{idx}:  {val}')
```
